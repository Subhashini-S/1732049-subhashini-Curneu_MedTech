```python
import pandas as pd
import numpy as np
import string
import re
user1 = pd.read_csv('D:\DATASETS/User1data.csv')
user1['Recommended'] 
user1.head() 
```

```python
user1['Selected'] = user1['Selected'].astype(str)
```

```python
y = []
for val in range(len(user1)) :
    data = user1.iloc[val,5]
    z = []
    for element in data:
        if str(element).isupper() == True:
            z.append(element)
            a = 1
        if str(element).isdigit() ==True and a ==1 :
            z.append(element)
    z=''.join(z)
    a=0
    y.append(z)
```

```python
y
```

```python
user1['Selected']=y
```
